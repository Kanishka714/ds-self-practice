```python
import pandas as pd
url = "https://raw.githubusercontent.com/selva86/datasets/master/Ionosphere.csv"
df = pd.read_csv(url)
df.head()
```

# **Seperating Independant and dependant Data**

```python
X = df.drop(["Class"], axis = 1)
Y = df["Class"]
print(X)
print(X.shape)
```

```python
print(Y)
```

# **Splitting data for training and testing**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

# **Import dependancies to work with ANN model**




```python
from keras.models import Sequential, load_model
from keras.layers import Dense, Dropout
```

# **Build Model**

```python
model = Sequential()
# Use X_train.columns if X_train is a Pandas DataFrame. Otherwise, use the appropriate method for accessing the column names
model.add(Dense(units = 10, activation = 'relu', input_dim = len(X_train.columns)))
model.add(Dense(units = 5, activation = 'relu'))
model.add(Dense(units = 1, activation = 'sigmoid'))
model.summary()
```

# **Compile the model**

```python
model.compile(loss = 'binary_crossentropy', optimizer= 'sgd', metrics= 'accuracy')
```

```python
model.fit(X_train, Y_train, validation_data = (X_test, Y_test) , epochs = 100, batch_size = 32)
```

```python
model.save('model1.h5')
```
