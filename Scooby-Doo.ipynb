```python
#(3)  Episode related data categorized and put into array

episode_details_columns = [
    'index', 'series.name', 'network', 'season', 'title', 'imdb', 'engagement', 'date.aired', 'run.time', 'format'
]
```

```python
episode_details_columns
```

```python
#(4)  checking for high imdb episodes

episode_details_df = df[episode_details_columns]
high_imdb_episodes = episode_details_df[episode_details_df["imdb"] > 7.5]
```

```python
high_imdb_episodes
```

```python
#(5)  checking the average run time of those episodes


mean_of_runtimes = high_imdb_episodes['run.time'].mean()
mean_of_runtimes
```

```python
less_than_mean = episode_details_df['run.time'] < mean_of_runtimes
```

```python
less_than_mean     #this is a boolean series
```

```python
#(6)  let's count how many (True)s are there.

count_less_than_mean = less_than_mean.sum()
count_less_than_mean
```

```python
#(7)  Getting a drataframe which shows episodes with more than 7.5 imdb and less than average runtime

less_time_high_gain = episode_details_df[(episode_details_df["run.time"] <= 24) & (episode_details_df["imdb"] > 7.5)]
```
